```python
img = cv2.imread('jl.webp')
```

```python
# 0 - 255 pixel range 
```

```python
cv2.imshow('myimage',img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
blur = cv2.blur(img,(10,10))
```

```python
def showImage(img_arr,winName = "my image",wait = 3000):
    cv2.imshow(winName,img_arr)
    cv2.waitKey(wait)
    cv2.destroyAllWindows()
```

```python
showImage(blur)
```

```python
edges = cv2.Canny(img , 400,550)
showImage(edges)
```

```python
# capturing vdo from web camera
```

```python
cap = cv2.VideoCapture(0)
while True:

    isPresent,frame = cap.read()
    cv2.imshow('myimage',frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    


cv2.destroyAllWindows()
cap.release()
```
