```python
from dputils.scrape import Scraper,Tag
```

```python
url ="https://www.flipkart.com/search?q=smart+tv+55+inch&sid=ckf%2Cczl&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_11_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_11_na_na_na&as-pos=1&as-type=RECENT&suggestionId=smart+tv+55+inch%7CTelevisions&requestId=45c4fa46-e6fd-42bf-8657-6b56833a2e82&as-searchtext=smart%20tv%2055"
scraper = Scraper(url)
```

```python
target_tag = Tag(name='div',cls="gdgoEp")
item_tag = Tag(name='div',cls='cPHDOP')
title_tag = Tag(name='div',cls='KzDlHZ',output='text')
price_tag = Tag(name='div',cls='Nx9bqj',output='text')
rating = Tag(name='div',cls='XQDdHH',output='text')
```

```python
products = scraper.get_repeating_data_from_page(
    target=target_tag,
    items=item_tag,
    title=title_tag,
    price=price_tag,
    rating=rating
)

for product in products:
    print(product)
```

```python
url1 = "https://www.flipkart.com/search?q=cosmic%20byte%20controller&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
scraper = Scraper(url1)
```

```python
target_tag = Tag(name='div',cls="_75nlfW")
item_tag = Tag(name='div',cls='slAVV4')
title_tag = Tag(name='a',cls='wjcEIp',output='text')
price_tag = Tag(name='div',cls='Nx9bqj',output='text')
link_tag = Tag(name='a',cls='wjcEIp',output='href')
img_tag = Tag(name='img',cls="DByuf4",output='src')
```

```python
productss = scraper.get_repeating_data_from_page(
    target=target_tag,
    items=item_tag,
    title=title_tag,
    price=price_tag,
    link=link_tag,
    img= img_tag
)

for product in productss:
    print(product)
```
