```python
from sqlalchemy import create_engine,Column,Integer,String
from sqlalchemy.orm import declarative_base,sessionmaker
```

```python
engine =  create_engine('sqlite:///database.db')
```

```python
base = declarative_base()
```

```python
class User(base):
    __tablename__ = "users"
    id = Column(Integer,primary_key=True)
    name = Column(String)
    email = Column(String)
    password = Column(String)
    
```

```python
base.metadata.create_all(engine)
```

```python
Session = sessionmaker(bind=engine)
session = Session()
```

```python
new_user = User(name= "Hood Hogan",email = "hogan.lko2025@gmail.com",password = "hogan123")
session.add(new_user)
session.commit()
```

```python
new_user = User(name = "yash",email= "yash2052@gmail.com",password = "123yash")
session.add(new_user)
session.commit()
```

```python
users = session.query(User).all()
users
```

```python
for u in users:
    print(f'ID:{u.id},name:{u.name},Password:{u.password}')
```

```python
users = session.query(User).filter_by(email= "yash2052@gmail.com").first()
print(users.name,users.password)
```

```python
users.name = "Farhan"
session.commit() 
```

```python
class Student(base):
    __tablename__ = "Student data"
    roll_no = Column(Integer,primary_key=True)
    name = Column(String)
    email = Column(String)
    course = Column(String)
    
```

```python
base.metadata.create_all(engine)
```

```python
Session = sessionmaker(bind=engine)
session = Session()
```

```python
new_user = Student(name= "Hood Hogan",email = "hogan.lko2025@gmail.com",course = "BBA")
session.add(new_user)
session.commit()
```

```python
new_user1 = Student(name= "Yash",email = "yash.lko2025@gmail.com",course = "BBA")
new_user2 = Student(name= "Ansh",email = "ansh.lko2025@gmail.com",course = "Btech")
new_user3 = Student(name= "Aadi",email = "aadi.lko2025@gmail.com",course = "BCA")
session.add(new_user1)
session.add(new_user2)
session.add(new_user3)
session.commit()
```

```python
data = session.query(Student).all()
data
```

```python
for d in data:
    print(f'roll_no:{d.roll_no},name:{d.name},email:{d.email},course:{d.course}')
```

```python
data = session.query(Student).filter_by(email= "harshit.lko2020@gmail.com").first()
print(data.name,data.course)
```
