```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.shape
```

```python
sns.histplot(df, x = 'age',bins=10,color='red')
```

```python
sns.histplot(df, x = 'age',bins=10,color='red',kde=True)
```

```python
import matplotlib.pyplot as plt
sns.histplot(df, x = 'age',bins=10,color='red',kde=True)
plt.xticks(range(0,81,5))
plt.show()
```

```python
sns.countplot(df, x = 'sex',palette='viridis')
plt.title('no of passengers by gender')
plt.show()
```

```python
sns.countplot(df, x = 'pclass', hue='sex')
plt.title('No of Passengers by class')
plt.show()
```

```python
sns.scatterplot(df, x ='age',y = 'fare',hue='sex',palette='magma')
plt.title('age v/s fare by gender')
plt.show()
```

```python
sns.scatterplot(df, x ='age',y = 'fare',hue='alive',size = 'pclass',palette='magma')
plt.title('age v/s fare by gender ')
plt.show()
```
